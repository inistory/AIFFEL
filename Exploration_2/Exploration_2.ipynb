## [Project 1] load_digits

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
digit = load_digits()

print(dir(digit))
```

```python
#Feature Data 지정하기
feature_data = digit.data

print(feature_data.shape) 
#shape는 배열의 형상정보를 출력
```

```python
#Label Data 지정하기
label_data = digit.target
print(label_data.shape)
label_data
```

```python
#Target Names 출력해 보기
digit.target_names
```

```python
#데이터 Describe 해 보기
print(digit.DESCR)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(feature_data, 
                                                    label_data, 
                                                    test_size=0.2, 
                                                    random_state=7)

print('X_train 개수: ', len(X_train), ', X_test 개수: ', len(X_test))
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
#SVM
from sklearn import svm
svm_model = svm.SVC()

random_forest = svm_model.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
#SGD Classifier
from sklearn.linear_model import SGDClassifier
sgd_model = SGDClassifier()

random_forest = sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
#Logistic Regression 
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()

random_forest = logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

모델 평가

- SVM 모델이  precision    recall  f1-score 이 균등하게 좋다
